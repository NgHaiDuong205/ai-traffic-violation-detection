```python
from pathlib import Path
import yaml

# Run this notebook with cwd = ai_modules/helmet_detect/training/
TRAINING_DIR = Path.cwd()
DATASET_ROOT = TRAINING_DIR / "dataset"  # put YOLO dataset here, or change path

data = {
    "path": str(DATASET_ROOT.resolve()),
    "train": "train/images",
    "val": "valid/images",
    "test": "test/images",
    "nc": 2,
    "names": ["helmet", "no helmet"],
}

data_yaml = TRAINING_DIR / "data.yaml"
with open(data_yaml, "w", encoding="utf-8") as f:
    yaml.dump(data, f, default_flow_style=False, allow_unicode=True)

print(f"data.yaml created at {data_yaml}")
```

```python
from pathlib import Path
from ultralytics import YOLO

TRAINING_DIR = Path.cwd()
data_yaml = TRAINING_DIR / "data.yaml"

model = YOLO("yolov8s.pt")

model.train(
    data=str(data_yaml),
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,  # GPU; use "cpu" if no CUDA
    workers=2,
    project=str(TRAINING_DIR / "runs"),
    name="helmet_detect",
)

# After training: copy best.pt -> ../weights/best3.pt for the backend API
```
